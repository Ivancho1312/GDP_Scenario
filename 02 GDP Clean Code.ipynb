# Опис на проект:

### Бруто-домашен производ (БДП)

- Бруто-домашен производ (БДП) (англиски: gross domestic product, GDP) — еден од основните показатели на големината и успехот на една економија. Бруто-домашниот производ е дефиниран како вкупната пазарна вредност на сите произведени финални добра и услуги во една земја во одреден период (вообичаено, една година). Според тоа, во БДП се вклучуваат само финалните добра, т.е. се исклучува вредноста на интермедијарите производи во производството. Изразот бруто се однесува на фактот дека амортизацијата на капиталните фондови е вклучена во пресметката на БДП. Значи, доколку се одземе амортизацијата на капиталните фондови би се добил нето домашен производ. Бруто-домашниот производ е централна макроекономска категорија преку која се мери вкупната економска активност на една земја. Во центарот на националните сметки е производството на стоките и услугите, односно бруто-домашниот производ (БДП), кој е главен показател за развиеноста на економијата на секоја држава. Со податокот за бруто-домашниот производ по жител се врши споредба на развиеноста на државата во однос на други држави, споредба на продуктивноста на трудот и истовремено податок од кој се добиваат податоците за бруто-националниот доход, бруто-штедењето и нето -задолжувањето на државата.

Бруто-домашен производ = Финална потрошувачка + бруто-инвестиции + извоз - увоз

### Финална потрошувачка

#### Трошоци за финална потрошувачка

- Трошоците за финална потрошувачка се составени од трошоци што ги прават резидентните институционални единици за стоки и за услуги што се употребуваат за директно задоволување на личните потреби или желби, како и на колективните потреби на членовите на заедницата. Трошоците за финална потрошувачка може да се направат на домашна територија или во странство.

Финална потрошувачка= Финална потрошувачка на домаќинствата + финална потрошувачка на НПИСД + финална потрошувачка на државата.

### Бруто-инвестиции

#### ИНВЕСТИЦИИ ВО ОСНОВНИ СРЕДСТВА

Инвестициите во основни средства се мерат преку вкупната вредност на набавените минус отуѓените основни средства од производителот во текот на пресметковниот период, плус некои унапредувања на вредноста на непроизведените средства реализирани преку производната активност на институционалната единица.

### Libraries:

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

---

### Барање 1:

```python
df = pd.read_csv('GDP_Data.csv')

df
```

---

### Барање 2:

```python
#Vadenje na kolona Year
df['Year'] = [s[:4] for s in df['Unnamed: 0']]

#Vadenje na colona Quarter
df['Quarter'] = [s[4:6] if s[4] in ('T', 'Т') else None for s in df['Unnamed: 0']]

df['Quarter'] = ['Q' + s[1] if s is not None else None for s in df['Quarter']]

#menuvanje od object vo float na vrednostite za ponatamosna obrabotka
regex = r'[^\d\.]'
cols_to_convert = ['Финална потрошувачка на домаќинствата, вклучувајќи НПИСД', 'Финална јавна потрошувачка', 'Бруто-инвестиции', 'Извоз на стоки и на услуги', 'Увоз на стоки и на услуги']
for col in cols_to_convert:
        df[col] = df[col].replace(regex, '', regex=True)
        df[col] = df[col].astype(float)


df['Year'] = df['Unnamed: 0'].str[:4].astype(int)


#Brisenje na coloni
df.drop(columns= ['Unnamed: 0.1',
       'Unnamed: 0'],inplace= True)

print(df.dtypes)
#Prefrlanje na kolonite Year i Quarter na pocetok na DataFramot
df = df.reindex(columns=['Year', 'Quarter'] + list(df.columns[:-2]))
#Gi dropnav kolonite ['Unnamed: 0.1'] i ['Unnamed: 0'] bidejki nemam potreba od tie podatoci, podatocite od ['Unnamed: 0'] gi raspredeliv vo drugi dve koloni Year i Quartal 
```

---

### Барање 3:

```python
#Dodavanja na nova kolona 'Бруто-домашен производ'
df['Бруто-домашен производ'] = df['Финална потрошувачка на домаќинствата, вклучувајќи НПИСД'] + df['Финална јавна потрошувачка'] + df['Бруто-инвестиции'] + df['Извоз на стоки и на услуги'] - df['Увоз на стоки и на услуги']
df
```

### Барање 4:

```python
data = df[df['Year'].isin([2020, 2021, 2022]) & df['Quarter'].isin(['Q1', 'Q2'])]

year_by_year = data.groupby('Year')['Бруто-домашен производ'].sum()
plt.figure(figsize=(15, 7))
plt.pie(year_by_year, labels=year_by_year.index, autopct='%1.2f%%', colors=['tomato', 'cornflowerblue', 'gold'], textprops={'fontsize':16},startangle=90)
plt.title('БДП за првите два квартали од 2020, 2021 и 2022 година', fontweight = 'bold', fontsize = 20)
plt.show()
```

```python
'''
Piechar:
Врз основа на дадените податоци за вредностите на БДП во 2020, 2021 и 2022 година, 
може да се заклучи дека со текот на времето има конзистентно и значително зголемување на БДП. 
Во 2020 година, БДП изнесува 29,6%, што се зголемува на 32,7% во 2021 година, а понатаму на 37,7% во првиот 
и вториот квартал од 2022 година. Овој податок сугерира дека економијата има добри резултати и бележи раст 
во изминатите неколку години. Зголемувањето на БДП може да се препише на различни фактори како што се владините политики, 
инвестициите во сектори на економијата и друго.
'''
```

### Барање 5:

```python
data = df[df['Year'].isin([2020, 2021, 2022])]
plt.figure(figsize=(15, 7))
sns.boxplot(x='Year', y='Бруто-домашен производ', hue='Quarter', data=data)
plt.title('Boxplots за БДП по квартал и година',fontsize=16)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel('Year', fontsize=18)
plt.ylabel('Бруто-домашен производ', fontsize=18)
plt.show()
```

```python
'''
Boxplot:
Boxplotot ја прикажува распределбата на БДП за годините 2020, 2021 и 2022 година во четири квартали. 
Во текот на секоја година, дистрибуцијата изгледа дека е прилично симетрична, при што средниот БДП 
се наоѓа во центарот на секој boxplot. Во различни квартали и години, распонот на вредноста на БДП 
се менува, што укажува на варијации во економскиот раст и нестабилноста.
Во некои квартови може да се забележат неколку многу високи бројки на БДП, особено на горната страна. 
Причините за овие исклучоци може да се должат на исклучителен економски успех или на други  околности во тоа време.
'''
```
